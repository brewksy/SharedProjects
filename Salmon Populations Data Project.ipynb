```python
# Import Data
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
# Temperature Data from Lupin station https://open.canada.ca/data/en/dataset/0baf6e3e-bd90-4893-8b54-4931e7c36ca5
path = "tia_2101303.txt"
stationData = pd.read_fwf(path, header=1)
stationData
```

```python
# Data cleanup
stationData.iloc[:, 0:1]
```

```python
#Column "year"/"sd"/"hd" are combined - we need to separate them first
stationDataFormat = stationData.iloc[:,0].str.split(" ", n=2, expand = True)
stationData.drop(stationData.columns[0], inplace=True, axis=1)
print(stationData)
print(stationDataFormat)
```

```python
#Add the new columns at the beginning, now separated
stationData.insert(0,"year",stationDataFormat.iloc[:,0])
print(stationData)
```

```python
# Now we add some descriptions to the columns
stationData.columns = ["Year", "Trop Nights 1", "Trop Nights 2", "Frost Days", "Consec Frost Days", "Ice Days", "Freeze Thaw Days", "Spring Freeze Thaw Days", "Growing Degree Days", "Cooling Degree Days", "Heating Degree Days", "Frost Free Season", "Frost Free Season Start", "Frost Free Season End", "Growing Season", "Growing Season Start", "Growing Season End"]
stationData
```

```python
#Looking to remove any row that has irrelevant data now.  The dataset uses "-9999" to indicate no data
FilterArray1 = stationData['Frost Days']!=-9999
stationData = stationData[FilterArray1]
```

```python
stationData
```

```python
#https://open.canada.ca/data/en/dataset/4107dc81-79c0-406e-a1e6-301509d9e506
path = "Kluane_NPR_Freshwater_Kokanee_Spawning_Counts_1976-2018_data.csv"
fishData = pd.read_csv(path, sep=",", encoding="latin-1", header=1)
fishData
```

```python
#Format the data to match the previous set of data
fishData = fishData.drop(columns="Identification")
fishData = fishData.drop(columns="Mois")
fishData = fishData.drop(columns="Jour")
fishData
```

```python
# Rename the columns and then find the totals of each year
fishData.columns = ["Year", "Fish Produced"]
fishDataFormat = fishData.groupby(['Year'])["Fish Produced"].sum()
```

```python
fishData = fishDataFormat.reset_index()
fishData
```

```python
print(fishData.shape)
print(stationData.shape)
```

```python
stationData = stationData.astype({"Year": int})
combinedData = pd.merge(left=stationData,right=fishData, left_on='Year', right_on='Year')
combinedData = combinedData.drop(columns=['Trop Nights 1', 'Trop Nights 2'])
print(combinedData.shape)
combinedData
```

```python
#First, let's find out the correlation between subsets
features = pd.DataFrame(combinedData, columns=["Year",
                                               "Frost Days",
                                               "Consec Frost Days",
                                               "Ice Days",
                                               "Freeze Thaw Days",
                                               "Spring Freeze Thaw Days",
                                               "Growing Degree Days",
                                               "Cooling Degree Days",
                                               "Heating Degree Days",
                                               "Frost Free Season",
                                               "Frost Free Season Start",
                                               "Frost Free Season End",
                                               "Growing Season",
                                               "Growing Season Start",
                                               "Growing Season End"
                                              ])
target = pd.DataFrame(combinedData, columns=["Fish Produced"])
```

```python
# Use a function to sort the array so it sorts the elements by the regressionModel score from least correlated
# to most correlated
def sortLast(val): 
    return val[-1] 
```

```python
#Use linear model to determine fit between each feature
import sklearn.linear_model as lm
import matplotlib.pyplot as plt

#build the model with x, y
accuracy = []
y = target["Fish Produced"]
for element in features:
    x = features[[element]]
    linearmodel = lm.LinearRegression()
    regressionModel = linearmodel.fit(x,y)
    predictedY = regressionModel.predict(x)
    plt.plot(x,predictedY)
    plt.scatter(x,y)
    plt.xlabel(element)
    plt.ylabel("Fish Produced")
    plt.show()
    accuracy.append((element, regressionModel.score(x,y)))
    

accuracy.sort(key = sortLast)  
accuracy
```

```python
#Now let's try to use Polynomials for the same singular elements
import sklearn.preprocessing as prep

y = target["Fish Produced"]
for element in features:
    x = features[[element]]
    print()
    print(element)
    for d in range(1,4):
        poly = prep.PolynomialFeatures(degree=d)
        print("Degree: {}".format(d))
        xPoly = poly.fit_transform(x)
        poly.fit(xPoly,y)
        lin = lm.LinearRegression()
        lin.fit(xPoly,y)
        print(lin.score(xPoly,y))
```

```python
# "Year" and "Ice Days" seem to have the highest correlation, and both at 3rd degree correlation
# Now let's look at multi-linear regressions
import statsmodels.api as sm
accuracy=[]
y = target["Fish Produced"]
for element1 in features:
    for element2 in features:
        if element1 == element2:
            pass
        else:
            x = features[[element1, element2]]
            regression = sm.OLS(y,x).fit()
            accuracy.append((element1, element2, regression.rsquared))

accuracy.sort(key = sortLast)  
accuracy
```

```python
accuracy=[]
y = target["Fish Produced"]
for element1 in features:
    for element2 in features:
        for element3 in features:
            if element1 == element2 or element1 == element3 or element2 == element3:
                pass
            else:
                x = features[[element1, element2, element3]]
                regression = sm.OLS(y,x).fit()
                accuracy.append((element1, element2, element3, regression.rsquared))

accuracy.sort(key = sortLast)  
accuracy
```

```python
### CONCLUSION ###

# The best accuracy I could find was multilinear with 2 variables:
# Ice Days and Heating Degree Days.
# This remains true, and shows an even stronger correlation with the erroneous datapoint removed

# While year has a better polynomial accuracy, I wanted to find data that was not related to year.

# From http://crd-data-donnees-rdc.ec.gc.ca/CDAS/products/IndicesTrends/Indices_Descriptions.pdf
# What this means:
# Ice Days: Number of days with Max-Temperature LESS THAN OR EQUAL to 0°C
# Heating Degree Days: The sum of degrees for days with a Mean-Temperature GREATER THAN 18°C
```

```python

# Let's deep dive on the data for this one:
y = target["Fish Produced"]
x = features[["Ice Days", "Heating Degree Days"]]
regression = sm.OLS(y,x).fit()
print(regression.summary())

# The formula is: (Ice Days * -141.3424) + (Heating Degree Days * 3.2062)
# So the amount of Fish Produced is negatively correlated by the number of Ice Days
## More Ice Days means LESS Fish Produced
# The amount of Fish Produced is positively correlated with Heating Degree Days
## More Degrees of days over 18C means MORE Fish Produced
```
